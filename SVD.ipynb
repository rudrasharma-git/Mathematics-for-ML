```python
import numpy as np
np.set_printoptions(precision = 4, suppress = True)
```

```python
a = np.array ([[1,2,3,4], [1,1,2,3], [0,1,1,0]])
```

```python
a
```

```python
u, s, vt = np.linalg.svd(a, full_matrices=True)
```

```python
sm
```

```python
S = np.zeros((3,4))
S
```

```python
S[:,:3] = sm
S
```

```python
A = (np.dot(np.dot(u,S),vt))
A
```

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image
import matplotlib.pyplot as plt
```

```python
url = "https://image2url.com/r2/default/images/1769428527194-8f7883cc-2159-463b-886b-ce90d2fe6914.jpg"
```

```python
IMG = io.imread(url)
```

```python
img = cv.cvtColor(IMG, cv.COLOR_BGR2GRAY)
```

```python
imgg = np.array(img, float)
```

```python
print(img)
print(imgg)
print(imgg.shape)
```

```python
plt.imshow(imgg, cmap = 'gray')
```

NOW IMAGE NEEDS TO BE SCALED SUCH THAT THE MEAN OF ALL VALUES BECOMES ZERO AND STD BECOMES 1

```python
imgg_scaled = (imgg - np.mean(imgg))/np.std(imgg)
```

```python
plt.imshow(imgg_scaled, cmap = 'gray')
```

```python
U , s, VT = np.linalg.svd(imgg_scaled)
```

```python
variancemap = np.round((s)/np.sum(s), decimals = 3)
```

```python
import seaborn as sns
sns.barplot(x = range(1, 51), y = variancemap[0:50], color = 'blue')
plt.tight_layout()
```

```python
print(len(variancemap))
```

Reconstructing Image with eigenvalues of majority importance, first 10 s's

```python
imp_s = 35
```

```python
re_img = np.dot(np.dot(U[:,:imp_s], np.diag(s[:imp_s])), VT[:imp_s,:])
```

```python
plt.imshow(re_img, cmap = 'gray')
```

WE OBSERVE THAT WITH S AS LESS AS 35 WE GET MOST OF THE INFORMATION NEEDED i.e. FACE IS RECOGNIZABLE
